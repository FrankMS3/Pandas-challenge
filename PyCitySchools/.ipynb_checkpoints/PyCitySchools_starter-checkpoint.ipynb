```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Getting the total number of schools, students and the total budget
total_schools = school_data_complete["school_name"].nunique()
total_students = school_data_complete["student_name"].count()
total_budget = school_data_complete["budget"].unique().sum()
# Getting the mean maths and reading scores
avg_mathscore = school_data_complete["maths_score"].mean()
avg_readingscore = school_data_complete["reading_score"].mean()
# Getting the % of students who passed (>=50) maths, reading, then both (overall)
pass_maths = school_data_complete[school_data_complete["maths_score"]>=50]
pctpass_maths = (len(pass_maths)/total_students) * 100
pass_reading = school_data_complete[school_data_complete["reading_score"]>=50]
pctpass_reading = (len(pass_reading)/total_students) * 100
pass_overall = school_data_complete[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)]
pctpass_overall = (len(pass_overall)/total_students) * 100
# Creating a DataFrame to hold these values
LGA_summary_df = pd.DataFrame({"Total Schools" : total_schools, 
                                "Total Students" : total_students,
                                "Total Budget" : total_budget, 
                                "Average Maths Score" : avg_mathscore, 
                                "Average Reading Score" : avg_readingscore,
                                "% Passing Maths" : pctpass_maths,
                                "% Passing Reading" : pctpass_reading,
                                "% Overall Passing" : pctpass_overall}, index=[0])
# Formatting data
LGA_summary_df["Total Students"] = LGA_summary_df["Total Students"].map("{:,}".format)
LGA_summary_df["Total Budget"] = LGA_summary_df["Total Budget"].map("${:,.2f}".format)

LGA_summary_df.head()
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Grouping school data by school name
school_data_complete = school_data_complete.set_index("school_name")
grouped_name_df = school_data_complete.groupby(["school_name"])
# Getting totals (and per student budget)
total_students_school = grouped_name_df.size()
perstudent_budget_school = grouped_name_df["budget"].mean()/grouped_name_df["size"].mean()
total_budget_school = perstudent_budget_school*total_students_school
# Getting average scores
avg_mathscore_school = grouped_name_df["maths_score"].mean()
avg_readingscore_school = grouped_name_df["reading_score"].mean()
# Getting percentage of students passing
pass_maths_school = school_data_complete.loc[school_data_complete["maths_score"]>=50].groupby(["school_name"]).size()
pass_reading_school = school_data_complete.loc[school_data_complete["reading_score"]>=50].groupby(["school_name"]).size()
pass_overall_school = school_data_complete.loc[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)].groupby(["school_name"]).size()
pctpass_maths_school = (pass_maths_school/total_students_school) * 100
pctpass_reading_school = (pass_reading_school/total_students_school) * 100
pctpass_overall_school = (pass_overall_school/total_students_school) * 100
# Creating a DataFrame to hold these values
school_summary_df = pd.DataFrame({"School Type" : grouped_name_df["type"].unique(),
                                    "Total Students" : total_students_school, 
                                    "Total School Budget" : total_budget_school, 
                                    "Per Student Budget" : perstudent_budget_school,
                                    "Average Maths Score" : avg_mathscore_school,
                                    "Average Reading Score" : avg_readingscore_school,
                                    "% Passing Maths" : pctpass_maths_school,
                                    "% Passing Reading" : pctpass_reading_school,
                                    "% Overall Passing" : pctpass_overall_school})
school_summary_df
```

```python
top_pctpass = school_summary_df.sort_values("% Overall Passing", ascending=False)
top_pctpass.head(5)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Getting average maths scores for each year level, and grouping by school
ninth_math = school_data_complete.loc[school_data_complete["year"]==9].groupby("school_name")["maths_score"].mean()
tenth_math = school_data_complete.loc[school_data_complete["year"]==10].groupby("school_name")["maths_score"].mean()
eleventh_math = school_data_complete.loc[school_data_complete["year"]==11].groupby("school_name")["maths_score"].mean()
twelfth_math = school_data_complete.loc[school_data_complete["year"]==12].groupby("school_name")["maths_score"].mean()
# Combining these values into a DataFrame
maths_scores = pd.DataFrame({"Year 9" : ninth_math,
                                "Year 10" : tenth_math,
                                "Year 11" : eleventh_math,
                                "Year 12" : twelfth_math})
maths_scores.index.name = "School Name"
maths_scores
```

## Reading Score by Year

* Perform the same operations as above for reading scores

```python
# Getting average reading scores for each year level, and grouping by school
ninth_reading = school_data_complete.loc[school_data_complete["year"]==9].groupby("school_name")["reading_score"].mean()
tenth_reading = school_data_complete.loc[school_data_complete["year"]==10].groupby("school_name")["reading_score"].mean()
eleventh_reading = school_data_complete.loc[school_data_complete["year"]==11].groupby("school_name")["reading_score"].mean()
twelfth_reading = school_data_complete.loc[school_data_complete["year"]==12].groupby("school_name")["reading_score"].mean()
# Combining these values into a DataFrame
reading_scores = pd.DataFrame({"Year 9" : ninth_reading,
                                "Year 10" : tenth_reading,
                                "Year 11" : eleventh_reading,
                                "Year 12" : twelfth_reading})
reading_scores.index.name = "School Name"
reading_scores
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Creating bins and labels
bins = [0, 585, 630, 645, 680]
groups = ["<$585", "$585-630", "$630-645", "$645-680"]
# Grouping data by these bins
school_summary_df["Spending Bins"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=groups)
grouped_spending_df = school_summary_df.groupby("Spending Bins")
grouped_spending_df = grouped_spending_df[["Average Maths Score", "Average Reading Score", "% Passing Maths", "% Passing Reading", "% Overall Passing"]].mean()
grouped_spending_df.index.name = "Per Student Budget"
grouped_spending_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Creating bins and labels
bins_size = [0, 1000, 2000, 5000]
groups_size = ["Small (<1000)", "Medium (1000-2000)", "Large (>2000)"]
# Grouping data by these bins
school_summary_df["School Size Bins"] = pd.cut(school_summary_df["Total Students"], bins_size, labels=groups_size)
grouped_size_df = school_summary_df.groupby("School Size Bins")
grouped_size_df = grouped_size_df[["Average Maths Score", "Average Reading Score", "% Passing Maths", "% Passing Reading", "% Overall Passing"]].mean()
grouped_size_df.index.name = "School Size (# of Students)"
grouped_size_df
```
